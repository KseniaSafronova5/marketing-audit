```python
import pandas as pd

def run_marketing_audit(file_path):
    # 1. –ó–∞–≥—Ä—É–∂–∞–µ–º –¥–∞–Ω–Ω—ã–µ
    df = pd.read_csv(file_path)
    
    # 2. –ë–µ—Ä–µ–º –¥–∞–Ω–Ω—ã–µ —Ç–æ–ª—å–∫–æ –∑–∞ –ø–æ—Å–ª–µ–¥–Ω–∏–π –¥–µ–Ω—å (–≤ —Ä–µ–∞–ª—å–Ω–æ—Å—Ç–∏ —ç—Ç–æ –±—É–¥–µ—Ç "–≤—á–µ—Ä–∞")
    # –î–ª—è —Ç–µ—Å—Ç–∞ –≤–æ–∑—å–º–µ–º –ø–æ—Å–ª–µ–¥–Ω—é—é –¥–∞—Ç—É –≤ —Ñ–∞–π–ª–µ
    last_day = df['–î–∞—Ç–∞'].max()
    yesterday_data = df[df['–î–∞—Ç–∞'] == last_day]
    
    alerts = []
    
    # 3. –ü—Ä–æ–≤–µ—Ä—è–µ–º –∫–∞–∂–¥—É—é –∫–∞–º–ø–∞–Ω–∏—é –Ω–∞ "–≤—à–∏–≤–æ—Å—Ç—å"
    for index, row in yesterday_data.iterrows():
        # –£—Å–ª–æ–≤–∏–µ 1: –°–ª–∏–≤ –±—é–¥–∂–µ—Ç–∞ –±–µ–∑ –ª–∏–¥–æ–≤
        if row['–ë—é–¥–∂–µ—Ç'] > 30 and row['–õ–∏–¥—ã'] == 0:
            alerts.append(f"üî¥ –ê–õ–ï–†–¢: –ö–∞–º–ø–∞–Ω–∏—è '{row['–ö–∞–º–ø–∞–Ω–∏—è']}' —Å–ª–∏–ª–∞ {row['–ë—é–¥–∂–µ—Ç']}$ –∏ –Ω–µ –ø—Ä–∏–Ω–µ—Å–ª–∞ –Ω–∏ –æ–¥–Ω–æ–≥–æ –ª–∏–¥–∞!")
        
        # –£—Å–ª–æ–≤–∏–µ 2: –ß—É–¥–æ–≤–∏—â–Ω–æ –Ω–∏–∑–∫–∏–π ROI
        elif row['ROI'] < -50:
            alerts.append(f"‚ö†Ô∏è –í–ù–ò–ú–ê–ù–ò–ï: –£ –∫–∞–º–ø–∞–Ω–∏–∏ '{row['–ö–∞–º–ø–∞–Ω–∏—è']}' –∫—Ä–∏—Ç–∏—á–µ—Å–∫–∏–π ROI: {row['ROI']}% (–ë—é–¥–∂–µ—Ç: {row['–ë—é–¥–∂–µ—Ç']}$, –í—ã—Ä—É—á–∫–∞: {row['–í—ã—Ä—É—á–∫–∞']}$)")

    # 4. –§–æ—Ä–º–∏—Ä—É–µ–º —Ñ–∏–Ω–∞–ª—å–Ω—ã–π –æ—Ç—á–µ—Ç
    if not alerts:
        return "‚úÖ –ú–∞—Ä–∫–µ—Ç–∏–Ω–≥ –≤ –Ω–æ—Ä–º–µ. –ö—Ä–∏—Ç–∏—á–µ—Å–∫–∏—Ö –∞–Ω–æ–º–∞–ª–∏–π –∑–∞ –≤—á–µ—Ä–∞ –Ω–µ –æ–±–Ω–∞—Ä—É–∂–µ–Ω–æ."
    else:
        return "\n".join(alerts)

# –¢–µ—Å—Ç–æ–≤—ã–π –∑–∞–ø—É—Å–∫
if __name__ == "__main__":
    report = run_marketing_audit('marketing_data.csv')
    print("--- –¢–ï–ö–°–¢ –î–õ–Ø TELEGRAM ---")
    print(report)
```
