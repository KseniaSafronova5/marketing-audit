```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

# Настройки
np.random.seed(42)
days = 30
campaigns = ['FB_Поиск_Черногория', 'Google_Аренда', 'IG_Релс_Промо', 'TikTok_Туризм']
start_date = datetime.now() - timedelta(days=days)

data = []

for i in range(days):
    current_date = (start_date + timedelta(days=i)).date()
    for camp in campaigns:
        spend = np.random.uniform(50, 150)
        clicks = int(spend * np.random.uniform(1.2, 2.5))
        leads = int(clicks * np.random.uniform(0.02, 0.05))
        revenue = leads * 120 * np.random.uniform(0.8, 1.2)
        
        # Инъекция аномалий
        if camp == 'TikTok_Туризм' and i in [10, 11, 12]:
            spend, leads, revenue = 250, 0, 0
        if i == 20:
            leads, revenue = 0, 0
        if camp == 'FB_Поиск_Черногория' and i == 25:
            spend, revenue = 400, 50

        data.append([current_date, camp, round(spend, 2), clicks, leads, round(revenue, 2)])

# Создаем DataFrame с русскими названиями
df = pd.DataFrame(data, columns=['Дата', 'Кампания', 'Бюджет', 'Клики', 'Лиды', 'Выручка'])

# Считаем классические метрики
df['CPL'] = (df['Бюджет'] / df['Лиды']).replace([np.inf, -np.inf], 0).fillna(0).round(2)
df['ROI'] = (((df['Выручка'] - df['Бюджет']) / df['Бюджет']) * 100).round(2)

df.to_csv('marketing_data.csv', index=False, encoding='utf-8-sig') # utf-8-sig чтобы Excel открыл красиво
print("✅ Файл 'marketing_data.csv' с русскими полями создан!")
```
